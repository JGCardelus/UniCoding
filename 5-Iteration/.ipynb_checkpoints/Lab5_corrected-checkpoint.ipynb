### ATM Simulator

```python
database = {}

n = int(input("Enter number of new customers: "))
for i in range(n):
    username = input("Enter customer username: ")
    savings = float(input("Enter customer savings: "))
    
    database[username] = savings
```

```python
user = None
while user == None:
    in_user = input("Introduce username please: ")

    for key in database:
        if key == in_user:
            user = key
            break

while True:
    print("Customer ID: %s // Savings: %s" % (user, database[user]))
    print("Menu options: ")
    print("    -- A) Withdraw money")
    print("    -- B) Deposting money")
    print("    -- C) Exit")
    print()
    print()
    
    menu_selection = input("    >>>  ")
    
    if menu_selection == "C":
        break
    elif menu_selection == "A":
        amount = float(input("How much money would you like to withdraw? "))
        if amount <= database[user]:
            print("Money withdrawn, printing...")
            database[user] -= amount
        else:
            print("Not enough money in account to perform transaction")
    elif menu_selection == "B":
        amount = float(input("How much money would you like to deposit? "))
        print("Contacting servers...")
        database[user] += amount
        print("Waiting for transaction verification...")
        print("Transaction key verified: 3RJ2LKJ392//&$JAKDJF")
        print("Transaction performed")
    
```

### Draw a Tic Tac Toe board

```python
import numpy as np

def draw_line(w, elems):
    for i in range(w):
        print(" ...", end="")
    print()
    for i in range(w):
        elem = elems[i]
        if elem == 1:
            elem = "O"
        elif elem == 0:
            elem = "X"
        else:
            elem = " "
        
        if i == 0:
            print("    | %s " % (elem), end="")
        else:
            print("| %s " % (elem), end="")
            
    print("|")
    
def draw_board(s, elems):
    if elems == None:
        elems = []
        for i in range(s):
            col = []
            for j in range(s):
                col.append(" ")
            elems.append(col)
    
    #Print row num
    for i in range(s):
        print(("{:5d}").format(i), end="")
    print()        
    for i in range(s):
        print(("{:2d}").format(i), end=". ")
        draw_line(s, elems[i])

def check_line(player, matrix):
    has_won = False
    for row in matrix:
        counter = 0
        for col in row:
            if col == player:
                counter += 1
          
        if counter == len(matrix[0]):
            has_won = True
            break
            
    return has_won
   
def check_diagonals(player, game):
    diagonal_a = []
    diagonal_b = []
    
    for i in range(len(game)):
        for j in range(len(game)):
            if i == j:
                diagonal_a.append(game[i][j])
            if len(game) - (i + 1) == j:
                diagonal_b.append(game[i][j])
                
    diagonal_a = np.array(diagonal_a)
    diagonal_b = np.array(diagonal_b)
                
    has_won = check_line(player, [diagonal_a, diagonal_b])
    return has_won
    
def check_game(player, game):
    game = np.array(game)
    
    has_won = check_line(player, game)
    if not has_won: has_won = check_line(player, np.transpose(game))
    if not has_won: has_won = check_diagonals(player, game)
    
    return has_won

def check_lost_state(game):
    lost = True
    for row in game:
        for col in row:
            if col == 3:
                lost = False
                
    return lost

def create_game(n):
    game = []
    for i in range(n):
        col = []
        for j in range(n):
            col.append(3)
        game.append(col)
        
    return game

def player_step(player, game):
    while True:
        x = int(input("X: "))
        y = int(input("Y: "))
        
        if x < len(game) and y < len(game):
            if game[y][x] == 3:
                game[y][x] = player
                print()
                break
            else:
                print("Those coordinates have already been selected")
        else:
            print("Those coordinates do not exist.")
    
def turn(player, identificator, game):    
    print("%s - Turn" % (identificator))
    player_step(player, game)
    has_won = check_game(player, game)
    if has_won == True:
        print("Player %s has won" % (identificator))
        return True
    draw_board(n, game)
    
    both_loose = check_lost_state(game)
    if both_loose:
        print("Sorry both players have lost.")
        return True
    
    return False
    
n = 3
game = create_game(n)
x_won = False
o_won = False

draw_board(n, game)

while True:
    end_play = turn(0, "X", game)
    if end_play: break
        
    end_play = turn(1, "0", game)
    if end_play: break
    
```
