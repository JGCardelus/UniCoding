```python
# VARIABLES

name = "Jorge" #String
surnames = ["Gonzalez", "Cardelus"] #List
email = "jgcardelus@gmail.com" #String
age = 18 #Int

#To check the type of a variable:
print(type(name))
print(type(surnames))
print(type(age))

#Some basic operations
full_name = name
for surname in surnames:
    full_name += " " + surname
    
print(full_name)
print("%s has %s years" % (name, str(age)))
```

```python
# IMPORT LIBS
import math
import numpy as np #np is the name which is always given to numpy
import matplotlib.pyplot as plt #plt is the name which is always given to matplotlib.pyplot

resultado = math.sqrt(2)
print((resultado, type(resultado)))

x = np.linspace(0, 2*np.pi, 100) #Creates from 0 to 2*pi a 100 digits equally apart
y = np.cos(x)  #It calculates the cosine of each value of x
plt.plot(x,y) #Plot y(x)
```
