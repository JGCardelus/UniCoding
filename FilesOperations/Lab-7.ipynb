### Get words from a long file

In this cell appart from reading the words, we are creating a dictionary with them and also counting how may times it appears so that only two for loops are required and there is no extra and unnecessary computing required, there for some statistics from 'Lab 7.5 - Some statistics from a text file' and 'Lab 7.6 - Frequency of words' are also computed here.

```python
# Using the file from last cell
words = []
dictionary = []
word_appearance = {}
n_chapters = 0


no_wanted_chars = """,;:.-_{}()"?¿!\@#$%&/[]"""
no_wanted_symbols = ["\n",""," "] #Maybe use in ASCII for more security

for line in lines:
    for raw_word in line.split(' '):
        word = ""
        for letter in raw_word:
            if letter not in no_wanted_chars:
                if letter not in no_wanted_symbols:
                    word += letter
        
        if len(word) == 0:
            continue
        
        words.append(word)
        if word not in dictionary:
            dictionary.append(word)
            
        if word == "Capítulo":
            n_chapters += 1
            
        if word not in word_appearance.keys():
            word_appearance[word] = 1
        else:
            n = word_appearance[word]
            word_appearance[word] = n + 1
```

```python
file = open('ElQuijote.txt', 'r', encoding='utf8')
print("File statistics")
n_chars = len(file.read())
n_lines = len(lines)
n_words = len(words)

total_length = 0
for word in dictionary:
    total_length += len(word)
n_average = total_length / len(dictionary)

print("  --> Number of characters: %s" % (n_chars))
print("  --> Number of lines: %s" % (n_lines))
print("  --> Number of words: %s" % (n_words))
print("  --> Number of chapters: %s" % (n_chapters))
print("  --> Average length of words: %s" % (n_average))
```

### Lab 7.7 Read binary files

```python
statistics_file = open('statistics-file.bin', 'rb')
raw_statistics = json.loads(statistics_file.read())

statistics = []
for key_rs in raw_statistics.keys():
    value_rs = raw_statistics[key_rs]
    if len(statistics) == 0:
        statistics.append(key_rs)
        continue
        
    for i, word in enumerate(statistics):
        value_word = raw_statistics[word]
        if value_rs >= value_word:
            statistics.insert(i, key_rs)
            break
        
for i in range(20):
    word = statistics[i]
    value = raw_statistics[word]
    print("word: '%s' -- frequency: %s" % (word, value))
```

### Use log files with tic tac toe

```python
import numpy as np
from datetime import datetime as dt

def log_start(file, n):
    file.write('%s Game begins. Game rows %s\n' % (dt.now(), n))
    
def log_end(file):
    file.write('%s Game over\n' % (dt.now()))
    
def log_game(file, game):
    file.write('%s - game: %s\n' % (dt.now(), game))

def log_winner(file, user):
    file.write('%s - winner: %s\n' % (dt.now(), user))
    
def log_move(file, user, x, y):
    file.write('%s - user-move - user: %s, x: %s, y: %s\n' % (dt.now(), user, x, y))

def draw_line(w, elems):
    for i in range(w):
        print(" ...", end="")
    print()
    for i in range(w):
        elem = elems[i]
        if elem == 1:
            elem = "O"
        elif elem == 0:
            elem = "X"
        else:
            elem = " "
        
        if i == 0:
            print("    | %s " % (elem), end="")
        else:
            print("| %s " % (elem), end="")
            
    print("|")
    
def draw_board(s, elems):
    if elems == None:
        elems = []
        for i in range(s):
            col = []
            for j in range(s):
                col.append(" ")
            elems.append(col)
    
    #Print row num
    print(("{:4s}").format(""), end="")
    for i in range(s):
        print(("{:4d}").format(i), end="")
    print()        
    for i in range(s):
        print(("{:2d}").format(i), end=". ")
        draw_line(s, elems[i])
        
#Checks if a certain row contains three elements which are the same
def check_line(player, matrix):
    has_won = False
    for row in matrix:
        counter = 0
        for col in row:
            if col == player:
                counter += 1
          
        if counter == len(matrix[0]):
            has_won = True
            break
            
    return has_won

#Creates a new matrix, containing both diagonals and the passes it
#to the check_line function which can calculate if there is a winnning
#state
def check_diagonals(player, game):
    diagonal_a = []
    diagonal_b = []
    
    for i in range(len(game)):
        for j in range(len(game)):
            if i == j:
                diagonal_a.append(game[i][j])
            if len(game) - (i + 1) == j:
                diagonal_b.append(game[i][j])
                
    diagonal_a = np.array(diagonal_a)
    diagonal_b = np.array(diagonal_b)
                
    has_won = check_line(player, [diagonal_a, diagonal_b])
    return has_won
    
#Checks the winning state for rows, columns and diagonals
def check_game(player, game):
    game = np.array(game)
    
    has_won = check_line(player, game)
    #The columns of the game matrix are the rows of the transposed
    #Therefore if we transposed the matrix it can be analysed by
    #check_line
    if not has_won: has_won = check_line(player, np.transpose(game))
    if not has_won: has_won = check_diagonals(player, game)
    
    return has_won

#Checks if all cells have been completed but if there is no 
#winning state
def check_lost_state(game):
    lost = True
    for row in game:
        for col in row:
            if col == 3:
                lost = False
                
    return lost

def turn(n, player, identificator, game, file):    
    print("%s - Turn" % (identificator))
    x, y = player_step(player, game)
    log_move(file, identificator, x, y)
    has_won = check_game(player, game)
    log_game(file, game)
    draw_board(n, game)
    if has_won == True:
        log_winner(file, identificator)
        print("Player %s has won" % (identificator))
        return True
    
    both_loose = check_lost_state(game)
    if both_loose:
        log_winner(file, 'tie')
        print("Sorry both players have lost.")
        return True
    
    return False

def get_input(game):
    try:
        correct = False
        x = int(input("X: "))
        y = int(input("Y: "))

        if x < len(game) and x >= 0:
            if y < len(game) and y >= 0:
                if game[y][x] == 3:
                    correct = True
                else:
                    print("Those coordinates have already been selected")
            else:
                print("The Y coordinates are invalid.")
        else:
            print("The X coordinates are invalid.")
    except ValueError:
        print("The input is not in the correct format, consider using numbers.")
        x = None
        y = None
        
    return correct, x, y

def player_step(player, game):
    x_final, y_final = None, None
    while True:
        correct, x, y = get_input(game)
        if correct:
            game[y][x] = player
            x_final, y_final = x, y
            print()
            break
            
    return x_final, y_final
            
def create_game(n):
    game = []
    for i in range(n):
        col = []
        for j in range(n):
            col.append(3)
        game.append(col)
        
    file = open('tictactoe-log.txt', 'a')
    return game, file

def game_over(file):
    log_end(file)
    file.close()
    reset = input("Do you want to play a new game? [Y/n]")
    if reset.lower() == 'n':
        print("Thank you for playing.")
    else:
        new_game(3)

def new_game(n):
    game, file = create_game(n)
    x_won = False
    o_won = False

    log_start(file, n)
    draw_board(n, game)

    while True:
        end_play = turn(n, 0, "X", game, file)
        if end_play: break

        end_play = turn(n, 1, "O", game, file)
        if end_play: break
            
    game_over(file)
```
