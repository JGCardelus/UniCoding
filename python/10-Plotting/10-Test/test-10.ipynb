The file contains 13 datasets, each identified by a word in the first column. Columns X and Y are the coordinates of each dataset point.

The objectives of the exercise are:

* Separate the 13 datasets and plot them as 13 scatter plots inside the same figure (use subplots) (6 points)
* Calculate the basic statistics for each dataset (mean(x), mean(y), standard_deviation(x), standard_deviation(y) and correlation(x,y)) and check that despite being 13 very different datasets, their statistics are practically the same. (2 points)
* Make a well-formatted printout in the form of a table (2 points)

TIP. Remember that the correlation coefficient is calculated as: np.corrcoef(x, y)[0][1]

```python
import pandas as pd
import numpy as np

something = pd.read_csv('DatasaurusDozen.tsv', delimiter='\t')
something.head()

class Coordinate:
    def __init__(self):
        self.x = []
        self.y = []
        
    def calc_stats(self, key):
        print("KEY: " + key)
        print()
        print("Mean X: " + str(np.mean(self.x)))
        print("Mean Y: " + str(np.mean(self.y)))
        
        print("STD X: " + str(np.std(self.x)))
        print("STD Y: " + str(np.std(self.y)))
        
        print("CORR X: " + str(np.corrcoef(self.x)))
        print("CORR Y: " + str(np.corrcoef(self.y)))

otherthings = {}
for i, line in enumerate(something["dataset"]):
    if line not in otherthings.keys():
        coord = Coordinate()
        coord.x.append(float(something["x"][i]))
        coord.y.append(float(something["y"][i]))
        otherthings[line] = coord
    else: 
        otherthings[line].x.append(float(something["x"][i]))
        otherthings[line].y.append(float(something["y"][i]))
        
print(len(otherthings.keys()))
```

```python
import matplotlib.pyplot as plt



fig, axes = plt.subplots(7, 2, figsize=(10,10))
counter_x = 0
counter_y = 0
for i, key in enumerate(otherthings.keys()):
    axes[counter_x, counter_y].scatter(otherthings[key].x, otherthings[key].y)
    counter_y += 1
    if counter_y == 2:
        counter_x += 1
        counter_y = 0
```

```python
for key in otherthings.keys():
    otherthings[key].calc_stats(key)
```
