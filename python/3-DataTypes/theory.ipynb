```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(0, 100, 100)
y = 2**(-x)

plt.plot(x, y)
plt.scatter(1, 0.6, s=100, label="Felicidad actual")
plt.scatter(2, 0.3, s=100, label="Felicidad mañana")
plt.title("Felicidad y notas de Ale en el curso")
plt.legend()
plt.show()
```

```python
y_prim = -1/(2**x) + 1
plt.plot(x, y_prim, label="Velocidad")
plt.plot(x, y, label="Atención")
plt.scatter(x[80], y_prim[80], label="Posición actual")
plt.ylabel("Velocidad")
plt.xlabel("Tiempo de clase")
plt.legend()
plt.title("Velocidad de Santiago en clase")
plt.show()
```
